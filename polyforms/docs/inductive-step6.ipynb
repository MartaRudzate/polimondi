# Polimondu bildes ar DrawScene

**1. piemērs:** Polimondu izkārtošana 1 rindā, ģeometriskās īpašības un trijstūru režģis.

```python
from polyforms.draw_scene import *
from polyforms.perfect_seq import *
from polyforms.polyiamond import Polyiamond
from polyforms.point_tg import *
import numpy as np

TH = np.sqrt(3)/2
storage = PerfectSeq()
myseq = storage.get_sequence('SEQ_4_3_A')

polygons = []
for i in range(3):
    N = len(myseq[i])
    print('seq = {}, len = {}'.format(myseq[i], N))
    polygon = Polyiamond(list(zip(range(N,0,-1), myseq[i])))
    polygons.append(polygon)

# scene = DrawScene(-4, -8, 80, 90, '#999999', 'solid', 0.25)
scene = DrawScene(-4, -4, 39, 47, '#999999', 'solid', 0.25)
scene.draw_polyiamond('p0', polygons[0], 'k', [0.0, 0.0], False)
scene.draw_polyiamond('p1', polygons[1], 'k', [18.0, 2.0], False)
#scene.draw_polyiamond('p2', polygons[2], 'k', [50.0,0.0])

scene.highlight('p1', [3, 7, 11, 14], '#DC143C')

#scene.highlight('p2', [3, 8, 13, 17], '#DC143C')
#scene.highlight('p2', [4, 9, 14, 18], '#2D9BF0')

print(polygons[0].get_rect_box())
print(polygons[1].get_rect_box())
print(polygons[2].get_rect_box())

scene.show_grid()
scene.set_size_in(14,10)
```
